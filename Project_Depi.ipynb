# Data Prepossing

### Fix Data Type 

```python
df["Order Date"]=pd.to_datetime(df["Order Date"],format="%d/%m/%Y")
df["Ship Date"]=pd.to_datetime(df["Ship Date"],format="%d/%m/%Y")
```

```python
df['Postal Code'] = df['Postal Code'].astype(str)
```

### Fill nan Values

```python
df[df["Postal Code"].isna()]#with same country , city and State 
```

```python
df["Postal Code"]=df["Postal Code"].fillna(5401)# by search on google
```

# feature engineering

```python
df["Shipping_Duration"] = (df["Ship Date"] - df["Order Date"]).dt.days
```

```python
df["Year_order"]=df["Order Date"].dt.year
df["Month_order"]=df["Order Date"].dt.strftime("%B")
df["Day_order_name"]=df["Order Date"].dt.strftime("%A")
df["Day_order_date"]=df["Order Date"].dt.day
df['Weekday_order_date']=df['Order Date'].dt.weekday
```

```python
df["Day_Ship_name"]=df["Ship Date"].dt.strftime("%A")
df["Month_Ship"]=df["Ship Date"].dt.strftime("%B")
df["Year_ship_date"]=df["Ship Date"].dt.year
df["Day_ship_date"]=df["Ship Date"].dt.day
df['Weekdayship_date']=df['Ship Date'].dt.weekday
```

# Questions on data

# What is the number of orders per year?

```python
sns.countplot(x='Year_order', data=df)
plt.title('Number of Orders per Year')
plt.xlabel('Order Year')
plt.ylabel('Number of Orders')
plt.show()
```

# Average sales per year?

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='Year_order', y='Sales', estimator='mean', palette='cividis')
plt.title('Average Sales per Year')
plt.ylabel('Average Sales')
plt.xlabel('Year')
plt.show()
```

# Which months have the highest number of orders?

```python
sns.countplot(x='Month_order', data=df, order=df['Month_order'].value_counts().index)
plt.title('Number of Orders per Month')
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()
```

# Which days of the week are most orders placed on?

```python
sns.countplot(x='Day_order_name', data=df, order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
plt.title('Orders by Day of the Week')
plt.xlabel('Day')
plt.ylabel('Number of Orders')
plt.show()
```

# Which days of the week are shipments made most often?

```python
sns.countplot(x='Day_Ship_name', data=df, order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
plt.title('Shipments by Day of the Week')
plt.xlabel('Day')
plt.ylabel('Number of Shipments')
plt.show()
```

# What is the most used shipping mode?

```python
sns.countplot(x='Ship Mode', data=df)
plt.title('Distribution of Shipping Modes')
plt.xlabel('Shipping Mode')
plt.ylabel('Count')
plt.show()
```

# Is there a relation between shipping mode and shipping duration?

```python
mean_duration = df.groupby('Ship Mode')['Shipping_Duration'].mean().reset_index()
sns.barplot(x='Ship Mode', y='Shipping_Duration', data=mean_duration)
plt.title('Average Shipping Duration by Shipping Mode')
plt.xlabel('Shipping Mode')
plt.ylabel('Average Days to Ship')
plt.show()
```

# Does shipping duration vary by city?

```python
top_cities = df['City'].value_counts().head(10).index
avg_duration = df[df['City'].isin(top_cities)].groupby('City')['Shipping_Duration'].mean().reset_index()
avg_duration = avg_duration.sort_values(by='Shipping_Duration', ascending=False)
sns.barplot(x='Shipping_Duration', y='City', data=avg_duration)
plt.title('Average Shipping Duration by Top 10 Cities')
plt.xlabel('Average Days to Ship')
plt.ylabel('City')
plt.show()
```

# Which customer segment places the most orders?

```python
sns.countplot(x='Segment', data=df)
plt.title('Orders per Customer Segment')
plt.xlabel('Segment')
plt.ylabel('Number of Orders')
plt.show()
```

# Average sales per segment ?

```python
avg_sales_segment = df.groupby('Segment')['Sales'].mean().reset_index()
sns.barplot(x='Segment', y='Sales', data=avg_sales_segment)
plt.title('Average Sales by Segment')
plt.xlabel('Segment')
plt.ylabel('Average Sales')
plt.show()
```

# Which region has the highest sales?

```python
region_sales = df.groupby('Region')['Sales'].sum().reset_index()
sns.barplot(x='Region', y='Sales', data=region_sales)
plt.title('Total Sales by Region')
plt.ylabel('Sales')
plt.xlabel('Region')
plt.show()
```

# What are the top-selling product categories?

```python
sns.barplot(x=df['Category'].value_counts().values, y=df['Category'].value_counts().index)
plt.title('Sales by Category')
plt.xlabel('Number of Orders')
plt.ylabel('Category')
plt.show()
```

# Which sub-categories generate the most sales?

```python
subcat_sales = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).reset_index()
sns.barplot(x='Sales', y='Sub-Category', data=subcat_sales)
plt.title('Total Sales by Sub-Category')
plt.xlabel('Sales')
plt.ylabel('Sub-Category')
plt.show()
```

# What are the top 10 most profitable cities (based on total sales)?

```python
top_sales_cities = df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10).reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=top_sales_cities, x='City', y='Sales', palette='crest')
plt.title('Top 10 Cities by Total Sales')
plt.xticks(rotation=45)
plt.show()
```

# Which shipping mode is preferred in each customer segment?

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Segment', hue='Ship Mode', palette='pastel')
plt.title('Shipping Mode Preference by Customer Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Number of Orders')
plt.legend(title='Ship Mode')
plt.show()
```

# Top customers by total sales

```python
top_customers = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10).reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=top_customers, x='Customer Name', y='Sales', palette='rocket')
plt.title('Top 10 Customers by Total Sales')
plt.xticks(rotation=45)
plt.show()
```

# Are there seasonal trends in order volume (by month & year)?

```python
orders_by_month_year = df.groupby(['Year_order', 'Month_order']).size().reset_index(name='Orders')
plt.figure(figsize=(12, 6))
sns.lineplot(data=orders_by_month_year, x='Month_order', y='Orders', hue='Year_order', palette='tab10')
plt.title('Monthly Order Trends by Year')
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.show()
```

# What are the most common combinations of segment and region?

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Region', hue='Segment', palette='Set2')
plt.title('Customer Segments per Region')
plt.xlabel('Region')
plt.ylabel('Number of Orders')
plt.legend(title='Segment')
plt.show()
```

# Which states have the slowest average shipping times?

```python
top_states = df.groupby('State')['Shipping_Duration'].mean().sort_values(ascending=False).head(10).reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=top_states, x='State', y='Shipping_Duration', palette='autumn')
plt.title('Top 10 States with Slowest Average Shipping Times')
plt.xticks(rotation=45)
plt.show()
```
